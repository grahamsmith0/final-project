```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df = df.dropna(subset=['neighborhood']) # get rid of rows without neighborhoods listed
# higher years delinquent (prior_years) and higher $ amount (current_delq), the worse the neighborhood is
# we can get rid of all the other columns as they aren't important
df = df.drop(
    columns=[
        'pin', 'billing_city', 'state_description', 'council_district', 'ward', 
        'tact', 'public_works_division', 'pli_division', 'police_zone', 'fire_zone', 'longitude', 'latitude'
    ])
```

```python
df.head(10)
```

```python
counts = {"neighborhood": [], "count": []}
for neighborhood in df["neighborhood"].unique():
    counts["neighborhood"].append(neighborhood)
    counts["count"].append(df.query(f"neighborhood == '{neighborhood}'").shape[0])

dfCounts = pd.DataFrame.from_dict(counts).sort_values("count")
dfCounts.head(10)
```

```python
# Drop neighborhoods with less than 10 accounts of tax deliqneuncy (Arlington Heights, Chateau, Glen Hazel, North Shore)
df = df.drop(df[df["neighborhood"].isin(["Arlington Heights", "Chateau", "Glen Hazel", "North Shore"])].index)
```

```python
dfNeighborhoodMedians = df.groupby(["neighborhood"]).median().reset_index()
dfNeighborhoodMediansDelq = dfNeighborhoodMedians.sort_values(by=['current_delq'], ascending=False)
dfNeighborhoodMediansYears = dfNeighborhoodMedians.sort_values(by=['prior_years'], ascending=False)
```

```python
dfNeighborhoodMediansDelq.plot.bar(x="neighborhood", y="current_delq", title="Neighborhoods sorted by median delinquent amount ($)", figsize=(20, 10))
```

```python
dfNeighborhoodMediansYears.plot.bar(x="neighborhood", y="prior_years", title="Neighborhoods sorted by median prior years delinquent", figsize=(20, 10))
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.scatter(dfNeighborhoodMedians["current_delq"], dfNeighborhoodMedians["prior_years"])

for i, neighborhood in enumerate(dfNeighborhoodMedians["neighborhood"]):
    ax.annotate(neighborhood, (dfNeighborhoodMedians["current_delq"][i], dfNeighborhoodMedians["prior_years"][i]), rotation=45)

ax.set_title("Median delinquent amount ($) vs. Median prior years delinquent")
ax.set_xlabel("Median delinquent amount ($)")
ax.set_ylabel("Median prior years delinquent")

    
plt.show()
```

Central Business District appears to be the most tax delinquent neighborhood, while Bluff appears to be the least.
