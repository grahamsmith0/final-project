# Introduction

Our goal was to determine the best neighborhood in Pittsburgh based on some datasets from the WRPDC. To figure out the best neighborhood, we reasoned that a neighborhood is only as good as its residents. In other words, we needed to come up with a metric to determine which neighborhood is the best based on how good its residents are. Eventually, we decided that a metric which would satisfy this condition would be `responsibility`. So, the neighborhood with the most responsible residents would be classified as the best neighborhood.

## The Metric & Datasets

 

The datasets we chose to explore the metric of responsibility are as follows:

 

- `Non-Traffic Citations`

 

- `Fire Incidents`

 

- `Tax Delinquency`

 

- `Neighborhoods With SNAP Data`

 

We used these datasets because we wanted to determine the level of responsibility of the neighborhoods based on three very different factors. <br>

One factor for determining the general responsibility of a neighborhood is the amount of non-traffic citations committed in that neighborhood. The rationale for using NTCs instead of more serious crimes is because these instances are less premeditated (such as public drunkeness or not having a correct license for something) so those who commit these crimes do so out of irresponsibility as opposed to malignance. <br>

Another factor was the amount of fire incidents in a neighborhood. Since a lot of these incidents are also mostly not from premeditated actions (most incidents were caused by cooking fires), this is also a good metric for showing how responsible someone is.<br>

Finally, its pretty self-explanatory that people who are tax delinquent are more irresponsible than those who aren't. <br><br>

In summary, neighborhoods with individuals who don't commonly commit public misdemeanors are most likely more responsible neighborhoods. Neighborhoods with residents who don't cause many fire incidents such as cooking fires will be more responsible. And neighborhoods with the least amount of tax delinquency will be the most responsible.<br>

The SNAP data was included for determining a basis for relating the datasets, or for giving contexts to each dataset.

# Non-Traffic Citations (Ezra)

`Non-traffic citations (NTCs)`, also known as summary offenses, are the most minor criminal offenses conducted in Pennsylvania. Examples range from disorderly conduct, loitering, and littering to animal abuse or illegal possession of fireworks.

In fact, according to the dataset, the five most common NTCs are:
> - Public Drunkenness (2687)
> - Harassment by Physical Contact, or Attempts or Threats (1311)
> - Any Person who urinates or defecates in, on, or about any public place (1100)
> - DISORDERLY CONDUCT - ENGAGE IN FIGHTING (1070)
> - Possessing a small amount of marijuana (1030)

For our metric of `"responsibility"`, neighborhoods with the **greatest amount** of NTCs are deemed the most irresponsible, while neighborhoods with the **least amount** of NTCs are deemed the most responsible

```python
data = pd.read_csv("Non-Traffic_Citations.csv") #non-traffic citations dataset
populations = pd.read_csv("Neighborhoods_with_SNAP_Data.csv") #dataset used for populations
```

```python
#getting rid of all columns except for offenses and neighborhoods
data = data.drop(columns=['PK', 'CCR', 'GENDER', 'RACE', 'AGE', 'CITEDTIME', 'INCIDENTLOCATION', 'ZONE', 'INCIDENTTRACT', 'COUNCIL_DISTRICT','PUBLIC_WORKS_DIVISION', 'X', 'Y'])
#Removing unimportant neighborhoods or neighborhoods not present in other datasets
data.dropna(subset="NEIGHBORHOOD",inplace=True) #getting rid of no neighborhood rows
data.drop(data[data["NEIGHBORHOOD"].isin(["Allegheny Center","Golden Triangle/Civic Arena","Outside City","Outside County","Outside State","Unable To Retrieve Address","Mt. Oliver Neighborhood","Mt. Oliver Boro"])].index, inplace=True)
data["NEIGHBORHOOD"].replace(to_replace="Mount Oliver",value="Mt. Oliver",inplace=True) #updating names for compatibility
data["NEIGHBORHOOD"].replace(to_replace="Central North Side",value="Central Northside",inplace=True)
```

However, the amount of summary offenses could be distorted based on the differences in size of the neighborhood populations. So, the best way to determine responsibility is to find the ratio of offenses to people in a neighborhood.

Unfortunately, the only dataset I could find with population data was last updated in 2010. Hopefully, it should still give an accurate representation of the differences in population between the neighborhoods.

```python
#Only need the 2010 population data for the neighborhoods from this dataset
populations = populations[["Neighborhood_2010_HOOD","Pop__2010"]]
#Removing neighborhoods so that the two datasets line up
populations.drop(populations[populations["Neighborhood_2010_HOOD"].isin(["Allegheny Center","Golden Triangle/Civic Arena","Outside City","Outside County","Outside State","Unable To Retrieve Address","Mt. Oliver Neighborhood","Mt. Oliver Boro"])].index, inplace=True)
#renaming the column for compatibility
populations.rename(columns = {'Neighborhood_2010_HOOD':'NEIGHBORHOOD'}, inplace = True)
#merging the NTC and population datasets
combined = pd.merge(data,populations,on="NEIGHBORHOOD",how="left")
```

```python
#Creating the ratio column
combined.insert(len(combined.columns),"RATIO","") #making a new empty column
#calculating the ratios
for i,rows in combined.iterrows():
    combined["RATIO"].iloc[i] = combined["OFFENSES"].iloc[i] / combined["Pop__2010"].iloc[i]
#sorting by ratio
combined.sort_values(by="RATIO",inplace=True)
#Removing neighborhoods with a population less than or equal to 100
combined.query("Pop__2010 > 100",inplace=True)
```

```python
display(combined)
```

#### Ratio Results

```python
combined.plot.bar(x="NEIGHBORHOOD",y="RATIO",figsize=(20,5),ylim=(0,1.0))
```

Now with population factored in, based on non-traffic citations, these are the **MOST** responsible neighborhoods (smallest ratio):
> 1. **Chartiers City**
> 2. Swisshelm Park
> 3. New Homestead
> 4. Stanton Heights
> 5. Ridgemont

#### Seeing the large number of Fire Incidences for each type of incident. We can only consider incidents that are more than 30 to keep the common incidents.

# Tax Delinquency (Graham)

```python
df = pd.read_csv("data.csv")
df = df.dropna(subset=['neighborhood']) # get rid of rows without neighborhoods listed
# higher years delinquent (prior_years) and higher $ amount (current_delq), the worse the neighborhood is
# we can get rid of all the other columns as they aren't important
df = df.drop(
    columns=[
        'pin', 'billing_city', 'state_description', 'council_district', 'ward', 
        'tact', 'public_works_division', 'pli_division', 'police_zone', 'fire_zone', 'longitude', 'latitude'
    ])
```

```python
df.head(10)
```

```python
counts = {"neighborhood": [], "count": []}
for neighborhood in df["neighborhood"].unique():
    counts["neighborhood"].append(neighborhood)
    counts["count"].append(df.query(f"neighborhood == '{neighborhood}'").shape[0])

dfCounts = pd.DataFrame.from_dict(counts).sort_values("count")
dfCounts.head(10)
```

```python
# Drop neighborhoods with less than 10 accounts of tax deliqneuncy (Arlington Heights, Chateau, Glen Hazel, North Shore)
df = df.drop(df[df["neighborhood"].isin(["Arlington Heights", "Chateau", "Glen Hazel", "North Shore"])].index)
```

```python
dfNeighborhoodMedians = df.groupby(["neighborhood"]).median().reset_index()
dfNeighborhoodMediansDelq = dfNeighborhoodMedians.sort_values(by=['current_delq'], ascending=False)
dfNeighborhoodMediansYears = dfNeighborhoodMedians.sort_values(by=['prior_years'], ascending=False)
```

```python
dfNeighborhoodMediansDelq.plot.bar(x="neighborhood", y="current_delq", title="Neighborhoods sorted by median delinquent amount ($)", figsize=(20, 10))
```

```python
dfNeighborhoodMediansYears.plot.bar(x="neighborhood", y="prior_years", title="Neighborhoods sorted by median prior years delinquent", figsize=(20, 10))
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.scatter(dfNeighborhoodMedians["current_delq"], dfNeighborhoodMedians["prior_years"])

for i, neighborhood in enumerate(dfNeighborhoodMedians["neighborhood"]):
    ax.annotate(neighborhood, (dfNeighborhoodMedians["current_delq"][i], dfNeighborhoodMedians["prior_years"][i]), rotation=45)

ax.set_title("Median delinquent amount ($) vs. Median prior years delinquent")
ax.set_xlabel("Median delinquent amount ($)")
ax.set_ylabel("Median prior years delinquent")

    
plt.show()
```

The only real pattern that emerges when comparing these two sets of data is that if a neighborhood has a high median amount of prior years delinquent, then they, on average, owe the least amount of money, and vice-versa. We decided as a group that the neighborhood with the least amount of years delinquent & least delinquent amount is the most responsible and the neighborhood with the most delinquent amount is the least responsible.

Squirrel Hill North appears to be the most tax delinquent neighborhood, while Bluff appears to be the least.
