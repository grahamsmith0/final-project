```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
## reading in the fire incidents dataset
fire_incidents = pd.read_csv("Fire_Incidents.csv")
```

```python
## dropping all fire incidents in which the neighborhood was not reported
fire_incidents = fire_incidents.dropna()
fire_incidents.head()
```

```python
## counting the number of fires for each neighborhood

unique = dict()

for index, row in fire_incidents.iterrows():
    
    neighborhoods = row["neighborhood"]
                       
    if neighborhoods not in unique.keys():
        unique[neighborhoods] = 1
    else:
        unique[neighborhoods] = unique[neighborhoods] + 1
        
unique1 = sorted(unique.items(), key=lambda x:x[1])
unique1
```

#### Seeing the large number of Fire Incidences for each neighborhood. I'd say that it is fair to drop the neighborhoods with less than 130 incidents.

```python
## thus, removing all the values more than 130 in the dictionary

for key, value in dict(unique).items():
    if value <= 130:
        del unique[key]
unique
```

```python
## counting the number of fires for each type of incident

occurences = dict()

for index, row in fire_incidents.iterrows():
    
    type = row["type_description"]
                       
    if type not in occurences.keys():
        occurences[type] = 1
    else:
        occurences[type] = occurences[type] + 1
        
occurences1 = sorted(occurences.items(), key=lambda x:x[1])
occurences1
```

#### Seeing the large number of Fire Incidences for each type of incident. We can only consider incidents that are more than 30 to keep the common incidents.

```python
## thus, removing all the type of incidents more than 30 in the dictionary

for key, value in dict(occurences).items():
    if value <= 30:
        del occurences[key]
occurences
```

```python
## plotting the number of incidents in each neighborhood using Matplotlib

types = list(occurences.keys())
value = list(occurences.values())

plt.xticks(rotation=90)
plt.rcParams["figure.figsize"] = (20,5.5)
plt.bar(range(len(occurences)), value, tick_label = types)
```
